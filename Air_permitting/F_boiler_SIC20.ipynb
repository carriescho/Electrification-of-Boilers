```python
mypath = 'C:\\Users\\zhang\\Box\\Air permit\\FL\\sic20'
outpath = 'C:\\Users\\zhang\\Box\\Air permit\\FL\\sic20_pdfs'
```

```python
from os import walk
zip_names = []
for (dirpath, dirnames, filenames) in walk(mypath):
    zip_names.extend(filenames)
```

```python
from zipfile import ZipFile
print('Extract single file from ZIP')
    # Create a ZipFile Object and load sample.zip in it
for zip in zip_names:
    zip_path = f'{mypath}/{zip}'
    with ZipFile( zip_path, 'r') as zipObj:
       # Get a list of all archived file names from the zip
        listOfFileNames = zipObj.namelist()
       # Iterate over the file names
        for fileName in listOfFileNames:
            print(fileName)
           # Check filename endsw
           # Extract a single file from zip
            zipObj.extract(fileName, outpath)
```

```python
import glob
import os
pdf_path = "sic20_pdfs/"
pdfs = glob.glob("{}/*.pdf".format(pdf_path))
```

```python
# importing required modules  
import PyPDF2  
    
having_boiler_pdfs = []
NotImplementedError_pdfs = []
for pdf in pdfs:
        # creating a pdf file object  
    pdfFileObj = open(pdf, 'rb')  
    try:
        # creating a pdf reader object  
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj)  

        # printing number of pages in pdf file  
        print(f'file: {pdf}; page numbers: {pdfReader.numPages}')
        page_nums = pdfReader.numPages
        having_bolier = False
        
        # creating a page object  for 1st page
        for pnum in range(page_nums):
            pageObj = pdfReader.getPage(pnum)  
            # extracting text from page  
            # print(pageObj.extractText())  

            contents = pageObj.extractText()
    #         print('*'*20)
            clist = contents.split(',')
            if not having_bolier:
                for line in clist:
                #     print(line)
                    ## check 'boiler' in eache line
                    if 'boiler' in line:
        #                 print(line, '\n')
                        having_bolier=True
    #                     print(f'pdf name: {pdf}')
                        having_boiler_pdfs.append(pdf)
    except:
        print('error: ', pdf)
        NotImplementedError_pdfs.append(pdf)
        pass
#         raise PdfReadError("File has not been decrypted")  
    finally:
        pdfFileObj.close()  
```

```python
import glob
import shutil
for f in pdfs:
    if f in having_boiler_pdfs:
        shutil.copy(f, r'C:\\Users\\zhang\\Box\\Air permit\\FL\\sic20_boiler')
```

```python
for f in pdfs:
    if f in NotImplementedError_pdfs:
        shutil.copy(f,r'C:\\Users\\zhang\\Box\\Air permit\\FL\\unextractable_sic20')
```
